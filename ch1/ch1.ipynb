## Random 
NumPyд№ҹжңүиЁұеӨҡж–№жі•дҫҶеүөе»әйҡЁж©ҹж•ёйҷЈеҲ—пјҡ

### rand
еүөе»әдёҖеҖӢзөҰе®ҡеҪўзӢҖзҡ„йҷЈеҲ—пјҢдёҰз”ЁдҫҶиҮӘж–ј[0, 1)зҡ„еқҮеӢ»еҲҶеёғзҡ„йҡЁж©ҹжЁЈжң¬еЎ«е……е®ғгҖӮ[[еҸғиҖғиіҮж–ҷ](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.rand.html)]

```python
np.random.rand(3)
```

```python
np.random.rand(5,5)
```

### randn

еҫһвҖңжЁҷжә–жӯЈж…ӢвҖқеҲҶдҪҲ[Пғ = 1]иҝ”еӣһжЁЈжң¬пјҲжҲ–жЁЈжң¬пјүгҖӮиҲҮ **rand** дёҚеҗҢпјҢйқ иҝ‘йӣ¶зҡ„еҖјжӣҙжңүеҸҜиғҪеҮәзҸҫгҖӮ[[еҸғиҖғиіҮж–ҷ](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.randn.html)]


жЁҷжә–жӯЈж…ӢеҲҶдҪҲжҳҜдёҖзЁ®зү№ж®Ҡзҡ„еёёж…ӢеҲҶдҪҲпјҢе…¶е№іеқҮж•ёпјҲОјпјүзӮә0пјҢжЁҷжә–е·®пјҲПғпјүзӮә1гҖӮеңЁжЁҷжә–жӯЈж…ӢеҲҶдҪҲдёӯпјҢж•ёж“ҡй»һдё»иҰҒйӣҶдёӯеңЁе№іеқҮеҖјйҷ„иҝ‘пјҢдёҰдё”йҡЁи‘—и·қйӣўе№іеқҮеҖјзҡ„еўһеҠ иҖҢе‘ҲжҢҮж•ёдёӢйҷҚгҖӮйҖҷзЁ®еҲҶдҪҲзҡ„жӣІз·ҡеҪўзӢҖе‘ҲйҗҳеҪўпјҢдёҰдё”еңЁе№іеқҮеҖјйҷ„иҝ‘жңүдёҖеҖӢе°–еі°гҖӮ

еңЁж©ҹеҷЁеӯёзҝ’е’ҢзөұиЁҲеӯёдёӯпјҢжЁҷжә–жӯЈж…ӢеҲҶдҪҲйҖҡеёёз”Ёж–је»әжЁЎе’ҢеҲҶжһҗпјҢдёҰдё”еңЁиЁұеӨҡз®—жі•е’ҢжҠҖиЎ“дёӯиў«е»ЈжіӣжҮүз”ЁгҖӮ

```python
nNp = np.random.randn(1000)
#print(nNp)
print(nNp.mean())
print(nNp.std())
```

### randint
еҫһ`low`пјҲеҢ…еҗ«пјүеҲ°`high`пјҲдёҚеҢ…еҗ«пјүиҝ”еӣһйҡЁж©ҹж•ҙж•ёгҖӮ [[еҸғиҖғиіҮж–ҷ](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html)]

```python
np.random.randint(1,100)
```

```python
np.random.randint(1,100,20)
```

```python
v2 = np.random.choice(range(1,101),10,replace=False)
#replace=False иЎЁзӨәдёҚе…ҒиЁұйҮҚиӨҮ
v2
```

```python
arr = np.arange(1,101)
print(arr)
np.random.shuffle(arr)
print(arr)
```

```python
import numpy as np 
print(np.random.rand(5))
np.random.seed(42)
print(np.random.rand(5))
```

```python
import numpy as np 
ran = np.random.default_rng(42)
print(ran.random(5))
print(ran.integers(1,100,10))
print(ran.shuffle([1,5,6,10]))
```
