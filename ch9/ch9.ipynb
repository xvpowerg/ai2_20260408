## 練習1:代數展開
公式：

$ (x + y)^3 $

$ x^{3} + 3 x^{2} y + 3 x y^{2} + y^{3} $

```python
import sympy as sp
from IPython.display import display,Math
x,y = sp.symbols("x y")
espr1 = (x + y) ** 3
expand_1 = sp.expand(espr1)
print(expand_1)
print(sp.latex(expand_1))
print(sp.latex(espr1))

mathObj = Math("x^{3} + 3 x^{2} y + 3 x y^{2} + y^{3}")
display(mathObj)
mathObj2 = Math(sp.latex(expand_1))
display(mathObj2)
```

### 練習2：因式分解
公式：
$[ x^3 + 3x^2 + 3x + 1 ]$

```python
x = sp.symbols("x")
expr2 = x**3 + 3*x**2 + 3*x + 1
f_expr2 = sp.factor(expr2)
print(f_expr2)
display(Math(sp.latex(f_expr2)))
```

### 練習3：代數簡化
公式：
$[ \frac{x^2 - y^2}{x - y} ]$

```python
x,y = sp.symbols("x y")
expr3 = (x**2 - y**2)/(x-y)
simp_expr3 = sp.simplify(expr3)
display(Math(sp.latex(simp_expr3))) 
```

### 練習4：符號代入
公式：
$[ x^2 + y^2 ]$
並代入 $( x = 1 )$ 和 $( y = 2 )$

```python
x,y = sp.symbols("x y")
expr4 = x ** 2 + y ** 2
print(type(x))
sub_expr4 = expr4.subs(x,1).subs(y,2)
print(sub_expr4)
```

### 練習5：求根
公式：
$[ x^2 - 4 = 0 ]$

```python
x = sp.symbols("x")
expor5 = x ** 2 - 4
sqrt_expr5 = sp.solve(expor5)
print(sqrt_expr5)
```
